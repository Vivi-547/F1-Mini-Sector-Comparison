```

import fastf1
from fastf1 import plotting
import matplotlib.pyplot as plt
import numpy as np

fastf1.Cache.enable_cache('cache')

def load_session(year, gp_name, session_type):
    session = fastf1.get_session(year, gp_name, session_type)
    session.load()
    return session

def plot_comparison(driver1, driver2, session):
    telemetry1 = session.laps.pick_driver(driver1).pick_fastest().get_telemetry()
    telemetry2 = session.laps.pick_driver(driver2).pick_fastest().get_telemetry()

    fig, ax = plt.subplots(figsize=(15, 8))
    ax.scatter(telemetry1['X'], telemetry1['Y'], c=telemetry1['Speed'], cmap='Reds', s=2, label=driver1)
    ax.scatter(telemetry2['X'], telemetry2['Y'], c=telemetry2['Speed'], cmap='Blues', s=2, label=driver2)

    ax.set_xlabel('X Position')
    ax.set_ylabel('Y Position')
    ax.legend()
    plt.show()

session = load_session(2024, 'Japan', 'Q')
plot_comparison('VER', 'HAM', session)
```
